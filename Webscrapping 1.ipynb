```python
#2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. Name, IMDB rating, Year of release) and save it in form of a CSV file.
page=requests.get("https://www.imdb.com/chart/top/")
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
title=soup.find_all('td',class_='titleColumn')
movie_title=[]
year_release = []
j=0
for i in range(0,100):
    mov = title[i].text.replace("\n",'')
    if(j < 9):
        num = 14
    else:
        num = 15
    mov_title = mov[num:]
    result = mov_title.index('(')
    movie_title.append(mov_title[0:result])
    year_release.append(mov_title[result+1:result+5])
    j += 1
```

```python
rating=soup.find_all('td',class_='ratingColumn imdbRating')
movie_rating=[]
for i in range(0,100):
    movie_rating.append(rating[i].text.replace("\n",''))
```

```python
import pandas as pd
IMDB_top_100=pd.DataFrame({})
IMDB_top_100['movies_name']=movie_title
IMDB_top_100['year_of_release']=year_release
IMDB_top_100['IMDB_rating']=movie_rating  
IMDB_top_100
```

```python
#Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content
#and the code for the video from the link for the youtube video from the post.
page=requests.get("https://coreyms.com/")
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
heading=[]
date=[]
content=[]
video=[]

for i in soup.find_all('a',class_='entry-title-link'):
    heading.append(i.text)
for i in soup.find_all('time',class_='entry-time'):
    date.append(i.text)
for i in soup.find_all('div',class_='entry-content'):
    content_p=i.find('p')
    content.append(content_p.text)
    try:
        video_src=i.find('span',class_='embed-youtube')
        link=video_src.find('iframe',class_='youtube-player')
        video.append(link['src'].split('"')[0].split("?")[0])
    except:
        video.append('NA')
        
```

```python
import pandas as pd
coreyms=pd.DataFrame({})
coreyms[' Heading']=heading
coreyms['Date']=date
coreyms['Content']=content
coreyms['Video']=video
coreyms
```

```python
#Write a python program to scrape first 10 product details which include product name , price , Image URL from
#https://www.bewakoof.com/women-tshirts?ga_q=tshirts .
url = "https://www.bewakoof.com/women-tshirts?ga_q=tshirts"
page14= requests.get(url)

#see page content
soup14= BeautifulSoup(page14.content, "html.parser")

Product_names= []  #emptylist
Price= []   #emptylist
URL_Image= []  #emptylist


#for Product_names
for i in soup14.find_all("div", class_="productCardDetail"):
    Product_names.append(i.text.split('₹')[0].strip())
    
#for Price
for i in soup14.find_all("span", class_="discountedPriceText"):
    Price.append(i.text.strip())
    
#for URL_Image
for i in soup14.find_all("img", class_="productImgTag"):
    URL_Image.append(i.get_attribute_list('src'))
    
# Make data frame
bewkoof=pd.DataFrame({})
bewkoof['Product_names']=Product_names
bewkoof['Price']=Price
bewkoof['URL_Image']=URL_Image
bewkoof
```
